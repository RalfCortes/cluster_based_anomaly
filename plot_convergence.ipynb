```python
import pickle
import plotly.graph_objects as go
import numpy as np
from plotly.subplots import make_subplots

N_cluster = 2
```

```python
with open(f'{N_cluster}_dict_opt_save_NOPCA_STD.pkl', 'rb') as f:
    loaded_dict = pickle.load(f)
        
with open(f'{N_cluster}_dict_opt_save_pso_NOPCA_STD.pkl', 'rb') as f:
    loaded_dict_pso = pickle.load(f)

with open(f'shape_{N_cluster}_dict_opt_save_NOPCA_STD.pkl', 'rb') as f:
    shape_loaded_dict = pickle.load(f)
        
with open(f'shape_{N_cluster}_dict_opt_save_pso_NOPCA_STD.pkl', 'rb') as f:
    shape_loaded_dict_pso = pickle.load(f)
```

```python
with open(f'{3}_dict_opt_save_NOPCA_STD.pkl', 'rb') as f:
    loaded_dict3 = pickle.load(f)
        
with open(f'{3}_dict_opt_save_pso_NOPCA_STD.pkl', 'rb') as f:
    loaded_dict_pso3 = pickle.load(f)

with open(f'shape_{3}_dict_opt_save_NOPCA_STD.pkl', 'rb') as f:
    shape_loaded_dict3 = pickle.load(f)
        
with open(f'shape_{3}_dict_opt_save_pso_NOPCA_STD.pkl', 'rb') as f:
    shape_loaded_dict_pso3 = pickle.load(f)
```

```python
temp_2_1 = np.array([shape_loaded_dict[i]["loss_sum"] for i in range(0, len(shape_loaded_dict))])
temp_2_2 = np.array([shape_loaded_dict_pso[i]["loss_sum"] for i in range(0, len(shape_loaded_dict_pso))])
temp_2_3 = np.array([loaded_dict[i]["loss_sum"] for i in range(0, len(loaded_dict))])
temp_2_4 = np.array([loaded_dict_pso[i]["loss_sum"] for i in range(0, len(loaded_dict_pso))])
```

```python
temp_3_1 = np.array([shape_loaded_dict3[i]["loss_sum"] for i in range(0, len(shape_loaded_dict3))])
temp_3_2 = np.array([shape_loaded_dict_pso3[i]["loss_sum"] for i in range(0, len(shape_loaded_dict_pso3))])
temp_3_3 = np.array([loaded_dict3[i]["loss_sum"] for i in range(0, len(loaded_dict3))])
temp_3_4 = np.array([loaded_dict_pso3[i]["loss_sum"] for i in range(0, len(loaded_dict_pso3))])
```

```python
temp_2_3[temp_2_3 >= 19000] = np.nan
temp_2_4[temp_2_4 >= 19000] = np.nan

temp_3_3[temp_3_3 >= 19000] = np.nan
temp_3_4[temp_3_4 >= 19000] = np.nan
```

```python
fig = make_subplots(
    rows=2,
    cols=2,
    subplot_titles=(
        f"N_cluster = {N_cluster}; Mode=Shape",
        f"N_cluster = {N_cluster}; Mode=Amplitude",
        f"N_cluster = {3}; Mode=Shape",
        f"N_cluster = {3}; Mode=Amplitude",
    ),
)

fig.add_trace(go.Scatter(y=temp_2_1, name="ga", marker_color="blue"), row=1, col=1)

fig.add_trace(
    go.Scatter(y=temp_2_2, name="pso", marker_color="red"),
    row=1,
    col=1,
)

fig.add_trace(
    go.Scatter(
        y=temp_2_3,
        name="ga",
        marker_color="blue",
        showlegend=False,
    ),
    row=1,
    col=2,
)

fig.add_trace(
    go.Scatter(
        y=temp_2_4,
        name="pso",
        marker_color="red",
        showlegend=False,
    ),
    row=1,
    col=2,
)

fig.add_trace(
    go.Scatter(
        y=temp_3_1,
        name="ga",
        marker_color="blue",
        showlegend=False,
    ),
    row=2,
    col=1,
)

fig.add_trace(
    go.Scatter(
        y=temp_3_2,
        name="pso",
        marker_color="red",
        showlegend=False,
    ),
    row=2,
    col=1,
)

fig.add_trace(
    go.Scatter(
        y=temp_3_3,
        name="ga",
        marker_color="blue",
        showlegend=False,
    ),
    row=2,
    col=2,
)

fig.add_trace(
    go.Scatter(
        y=temp_3_4,
        name="pso",
        marker_color="red",
        showlegend=False,
    ),
    row=2,
    col=2,
)

fig.update_layout(
    title=f"Loss function convergence",
    xaxis_title="Iteration",
    yaxis_title="Loss",
    margin=dict(l=50, r=50, t=50, b=20),
)
```
